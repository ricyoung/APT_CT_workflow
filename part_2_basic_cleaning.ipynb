First lets clean up on the status ( col 15 - overall status)
removing 
Suspended,Not yet recruiting, Withdrawn

```python
# df_filtered = cleaning[cleaning['OverallStatus'] == 'Withdrawn' ]

# cleaning.drop(df.index[df['Col1'] == 0], inplace=True)
cleaning.drop(cleaning.index[cleaning['OverallStatus'] == ('Withdrawn')], inplace=True)

# cleaning.drop(cleaning.index[cleaning['OverallStatus'] == ('Terminated')], inplace=True)

cleaning.drop(cleaning.index[cleaning['OverallStatus'] == ('Not yet recruiting')], inplace=True)

cleaning.drop(cleaning.index[cleaning['OverallStatus'] == ('Suspended')], inplace=True)

cleaning.drop(cleaning.index[cleaning['OverallStatus'] == ('Unknown status')], inplace=True)
```

Now since we are looking at enrollment - lets remove all the studies that do not have at least one enrolled person

```python

cleaning.drop(cleaning.index[cleaning['EnrollmentCount'] < 1 ], inplace=True)

# remove null values

cleaning = cleaning[cleaning['EnrollmentCount'].notna()]
```

```python
# convert to int
cleaning = cleaning.astype({'EnrollmentCount': 'int32'})
# cleaning = cleaning[cleaning['EnrollmentCount'].notna()]
```
